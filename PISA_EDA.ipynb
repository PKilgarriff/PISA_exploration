### Function that creates a database connection for a given country code

```python
def generate_country_database_connection(country_code):
  # Postgres username, password, and database name
  POSTGRES_ADDRESS = f"seta-{country_code}.cvcpj1fhj3k9.us-east-2.rds.amazonaws.com" ## INSERT YOUR DB ADDRESS IF IT'S NOT ON PANOPLY
  POSTGRES_PORT = '5432'
  POSTGRES_USERNAME = 'seta'
  POSTGRES_PASSWORD = 'defaultUnsafePassword'
  POSTGRES_DBNAME = country_code ## The country code that the database is named after
  # A long string that contains the necessary Postgres login information
  postgres_str = ('postgresql://{username}:{password}@{ipaddress}:{port}/{dbname}'
    .format(
      username=POSTGRES_USERNAME,
      password=POSTGRES_PASSWORD,
      ipaddress=POSTGRES_ADDRESS,
      port=POSTGRES_PORT,
      dbname=POSTGRES_DBNAME
      )
    )
  # Create the connection
  cnx = create_engine(postgres_str)
  return cnx
```

Create the PostgreSQL connections to each db

```python
country_connections = {}
for country_code in country_codes:
  print(f"Generating Connection for {country_code}")
  country_connections[country_code] = generate_country_database_connection(country_code)
```

Create the DataFrames for each database
> This takes a while as all columns are being selected and it'll only get worse as more entries come in

> The below code takes ~3mins30s to run, don't run unless necessary

```python
country_dataframes = {}
for country_code in country_codes:
  print(country_code)
  country_dataframes[country_code] = pd.read_sql_query('''SELECT * FROM responses;''', country_connections[country_code])
```
